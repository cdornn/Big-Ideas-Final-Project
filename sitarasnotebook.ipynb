```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

%matplotlib inLine

nonTracficCitation = pd.read_csv('nonTracficCitation.csv')

nonTracficCitation  = nonTracficCitation [nonTracficCitation ['INCIDENTTRACT'].notna()]

neighborhood = nonTracficCitation['NEIGHBORHOOD'].head(12)

INCIDENTTRACT = nonTracficCitation['INCIDENTTRACT'].head(12)

fig = plt.figure(figsize =(20, 20))

plt.bar(neighborhood[0:5], INCIDENTTRACT[0:5],color ='maroon', width = 0.9)

plt.title("Top 4 Neighborhood Non Traffic Incidence Track ")

plt.xlabel("NEIGHBORHOOD")

plt.ylabel("INCIDENTTRACT")

plt.show()

 
```

## Conclusion 

As you can see on the graph North Oakland is the best neighborhood too live in and the worst of the Neighborhood to live in Upper Lawerenceville and the best is North Oakland. This is based on the Non Traffic Citation. This means citations like speeding tickets and DUI aren't included in this spesific data set.
